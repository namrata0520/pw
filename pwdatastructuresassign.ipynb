```python


### 1. Reverse a String

def reverse_string(s):
 return s[::-1]

# Test
print(reverse_string("hello")) # Output: "olleh"


### 2. Count the Number of Vowels in a String

def count_vowels(s2):
 vowels = "aeiouAEIOU"
 return sum(1 for char in s2 if char in vowels)

# Test
print(count_vowels("hello")) # Output: 2


### 3. Check if a Given String is a Palindrome

def is_palindrome(s):
 return s == s[::-1]

# Test
print(is_palindrome("madam")) # Output: True


### 4. Check if Two Given Strings are Anagrams

def are_anagrams(s1, s2):
 return sorted(s1) == sorted(s2)

# Test
print(are_anagrams("listen", "silent")) # Output: True


### 5. Find All Occurrences of a Substring within Another String

def find_substring_occurrences(s, sub):
 return [i for i in range(len(s)) if s.startswith(sub, i)]

# Test
print(find_substring_occurrences("ababab", "ab")) # Output: [0, 2, 4]


### 6. Perform Basic String Compression Using Counts of Repeated Characters

def compress_string(s):
 compressed = []
 count = 1
 for i in range(1, len(s)):
   if s[i] == s[i-1]:
     count += 1

### 1. Reverse a String

def reverse_string(s):
    return s[::-1]

# Test
print(reverse_string("hello")) # Output: "olleh"


### 2. Count the Number of Vowels in a String

def count_vowels(s):
    vowels = "aeiouAEIOU"
    return sum(1 for char in s if char in vowels)

# Test
print(count_vowels("hello")) # Output: 2


### 3. Check if a Given String is a Palindrome

def is_palindrome(s):
    return s == s[::-1]

# Test
print(is_palindrome("madam")) # Output: True


### 4. Check if Two Given Strings are Anagrams

def are_anagrams(s1, s2):
    return sorted(s1) == sorted(s2)

# Test
print(are_anagrams("listen", "silent")) # Output: True


### 5. Find All Occurrences of a Substring within Another String

def find_substring_occurrences(s, sub):
    return [i for i in range(len(s)) if s.startswith(sub, i)]

# Test
print(find_substring_occurrences("ababab", "ab")) # Output: [0, 2, 4]


### 6. Perform Basic String Compression Using Counts of Repeated Characters

def compress_string(s):
    compressed = []
    count = 1
    for i in range(1, len(s)):
        if s[i] == s[i-1]:
            count += 1
        else:
            compressed.append(s[i-1] + str(count))
            count = 1
    compressed.append(s[-1] + str(count)) # Add the last character
    return "".join(compressed)

# Test
print(compress_string("aaabbcccc")) # Output: "a3b2c4"


### 7. Determine if a String has All Unique Characters

def has_unique_characters(s):
    return len(s) == len(set(s))

# Test
print(has_unique_characters("abcde")) # Output: True


### 8. Convert a String to Uppercase or Lowercase

def convert_case(s, to_upper=True):
    return s.upper() if to_upper else s.lower()

# Test
print(convert_case("hello", False)) # Output: "hello"


### 9. Count the Number of Words in a String

def count_words(s):
    return len(s.split())

# Test
print(count_words("Hello, how are you?")) # Output: 4


### 10. Concatenate Two Strings without Using the ‘+‘ Operator

def concatenate_strings(s1, s2):
    return "".join([s1, s2])

# Test
print(concatenate_strings("hello", "world")) # Output: "helloworld"


### 11. Remove All Occurrences of a Specific Element from a List

def remove_element(lst, element):
    return [item for item in lst if item != element]

# Test
print(remove_element([1, 2, 3, 4, 3], 3)) # Output: [1, 2, 4]


### 12. Find the Second Largest Number in a Given List of Integers

def second_largest(lst):
    unique_lst = list(set(lst)) # Remove duplicates
    unique_lst.sort()
    return unique_lst[-2] if len(unique_lst) > 1 else None

# Test
print(second_largest([1, 2, 3, 4, 5])) # Output: 4


### 13. Count the Occurrences of Each Element in a List and Return a Dictionary

from collections import Counter

def count_occurrences(lst):
    return dict(Counter(lst))

# Test
print(count_occurrences([1, 2, 2, 3, 3, 3])) # Output: {1: 1, 2: 2, 3: 3}


### 14. Reverse a List In-Place Without Using Built-In Reverse Functions

def reverse_list(lst):
    left, right = 0, len(lst) - 1
    while left < right:
        lst[left], lst[right] = lst[right], lst[left]
        left += 1
        right -= 1

# Test
lst = [1, 2, 3, 4]
reverse_list(lst)
print(lst) # Output: [4, 3, 2, 1]


### 15. Find and Remove Duplicates from a List While Preserving the Original Order of Elements

def remove_duplicates(lst):
    seen = set()
    return [item for item in lst if item not in seen and not seen.add(item)]

# Test
print(remove_duplicates([1, 2, 2, 3, 4, 3])) # Output: [1, 2, 3, 4]


### 16. Check if a List is Sorted (Either in Ascending or Descending Order)

def is_sorted(lst):
    return lst == sorted(lst) or lst == sorted(lst, reverse=True)

# Test
print(is_sorted([1, 2, 3])) # Output: True
print(is_sorted([3, 2, 1])) # Output: True
print(is_sorted([1, 3, 2])) # Output: False


### 17. Merge Two Sorted Lists into a Single Sorted List

def merge_sorted_lists(lst1, lst2):
    return sorted(lst1 + lst2)

# Test
print(merge_sorted_lists([1, 3, 5], [2, 4, 6])) # Output: [1, 2, 3, 4, 5, 6]


### 18. Find the Intersection of Two Given Lists

def intersection(lst1, lst2):
    return list(set(lst1) & set(lst2))

# Test
print(intersection([1, 2, 3], [3, 4, 5])) # Output: [3]


### 19. Find the Union of Two Lists Without Duplicates

def union(lst1, lst2):
    return list(set(lst1) | set(lst2))

# Test
print(union([1, 2, 3], [3, 4, 5])) # Output: [1, 2, 3, 4, 5]


### 20. Shuffle a List Randomly Without Using Built-In Shuffle Functions

import random

def shuffle_list(lst):
    random.shuffle(lst)
    return lst

# Test
print(shuffle_list([1, 2, 3, 4])) # Output: Randomly shuffled list


### 21. Find Common Elements in Two Tuples

def common_elements(tup1, tup2):
    return tuple(set(tup1) & set(tup2))

# Test
print(common_elements((1, 2, 3), (3, 4, 5))) # Output: (3,)


### 22. Find the Intersection of Two Sets of Integers

def set_intersection(set1, set2):
    return set1 & set2

# Test
set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(set_intersection(set1, set2)) # Output: {3}


### 23. Concatenate Two Tuples

def concatenate_tuples(tup1, tup2):
    return tup1 + tup2

# Test
print(concatenate_tuples((1, 2), (3, 4))) # Output: (1, 2, 3, 4)


### 24. Find the Union of Two Sets of Strings

def set_union(set1, set2):
    return set1 | set2

# Test
set1 = {"apple", "banana"}
set2 = {"banana", "cherry"}
print(set_union(set1, set2)) # Output: {'banana', 'apple', 'cherry'}


### 25. Find the Maximum and Minimum Values in a Tuple Using Tuple Unpacking

def max_min_tuple(tup):
    max_val, min_val = max(tup), min(tup)
    return max_val, min_val

# Test
print(max_min_tuple((5, 2, 9, 1))) # Output: (9, 1)


### 26. Union, Intersection, and Difference of Two Sets

def set_operations(set1, set2):
    union = set1 | set2
    intersection = set1 & set2
    difference = set1 - set2
    return union, intersection, difference

# Test
print(compress_string("aaabbcccc")) # Output: "a3b2c4"


### 7. Determine if a String has All Unique Characters

def has_unique_characters(s):
 return len(s) == len(set(s))

# Test
print(has_unique_characters("abcde")) # Output: True


### 8. Convert a String to Uppercase or Lowercase

def convert_case(s, to_upper=True):
 return s.upper() if to_upper else s.lower()

# Test
print(convert_case("hello", False)) # Output: "hello"


### 9. Count the Number of Words in a String

def count_words(s):
 return len(s.split())

# Test
print(count_words("Hello, how are you?")) # Output: 4


### 10. Concatenate Two Strings without Using the ‘+‘ Operator

def concatenate_strings(s1, s2):
 return "".join([s1, s2])

# Test
print(concatenate_strings("hello", "world")) # Output: "helloworld"


### 11. Remove All Occurrences of a Specific Element from a List

def remove_element(lst, element):
 return [item for item in lst if item != element]

# Test
print(remove_element([1, 2, 3, 4, 3], 3)) # Output: [1, 2, 4]


### 12. Find the Second Largest Number in a Given List of Integers
def second_largest(lst):
 unique_lst = list(set(lst)) # Remove duplicates
 unique_lst.sort()
 return unique_lst[-2] if len(unique_lst) > 1 else None

# Test
print(second_largest([1, 2, 3, 4, 5])) # Output: 4


### 13. Count the Occurrences of Each Element in a List and Return a Dictionary

from collections import Counter

def count_occurrences(lst):
 return dict(Counter(lst))

# Test
print(count_occurrences([1, 2, 2, 3, 3, 3])) # Output: {1: 1, 2: 2, 3: 3}


### 14. Reverse a List In-Place Without Using Built-In Reverse Functions

def reverse_list(lst):
 left, right = 0, len(lst) - 1
 while left < right:
  lst[left], lst[right] = lst[right], lst[left]
  left += 1
  right -= 1

# Test
lst = [1, 2, 3, 4]
reverse_list(lst)
print(lst) # Output: [4, 3, 2, 1]


### 15. Find and Remove Duplicates from a List While Preserving the Original Order of Elements

def remove_duplicates(lst):
 seen = set()
 return [item for item in lst if item not in seen and not seen.add(item)]

# Test
print(remove_duplicates([1, 2, 2, 3, 4, 3])) # Output: [1, 2, 3, 4]


### 16. Check if a List is Sorted (Either in Ascending or Descending Order)

def is_sorted(lst):
 return lst == sorted(lst) or lst == sorted(lst, reverse=True)

# Test
print(is_sorted([1, 2, 3])) # Output: True
print(is_sorted([3, 2, 1])) # Output: True
print(is_sorted([1, 3, 2])) # Output: False


### 17. Merge Two Sorted Lists into a Single Sorted List

def merge_sorted_lists(lst1, lst2):
 return sorted(lst1 + lst2)

# Test
print(merge_sorted_lists([1, 3, 5], [2, 4, 6])) # Output: [1, 2, 3, 4, 5, 6]


### 18. Find the Intersection of Two Given Lists

def intersection(lst1, lst2):
 return list(set(lst1) & set(lst2))

# Test
print(intersection([1, 2, 3], [3, 4, 5])) # Output: [3]


### 19. Find the Union of Two Lists Without Duplicates

def union(lst1, lst2):
 return list(set(lst1) | set(lst2))

# Test
print(union([1, 2, 3], [3, 4, 5])) # Output: [1, 2, 3, 4, 5]


### 20. Shuffle a List Randomly Without Using Built-In Shuffle Functions

import random

def shuffle_list(lst):
 random.shuffle(lst)
 return lst

# Test
print(shuffle_list([1, 2, 3, 4])) # Output: Randomly shuffled list


### 21. Find Common Elements in Two Tuples

def common_elements(tup1, tup2):
 return tuple(set(tup1) & set(tup2))

# Test
print(common_elements((1, 2, 3), (3, 4, 5))) # Output: (3,)


### 22. Find the Intersection of Two Sets of Integers

def set_intersection(set1, set2):
 return set1 & set2

# Test
set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(set_intersection(set1, set2)) # Output: {3}


### 23. Concatenate Two Tuples

def concatenate_tuples(tup1, tup2):
 return tup1 + tup2

# Test
print(concatenate_tuples((1, 2), (3, 4))) # Output: (1, 2, 3, 4)


### 24. Find the Union of Two Sets of Strings

def set_union(set1, set2):
 return set1 | set2

# Test
set1 = {"apple", "banana"}
set2 = {"banana", "cherry"}
print(set_union(set1, set2)) # Output: {’banana’, ’apple’, ’cherry’}


### 25. Find the Maximum and Minimum Values in a Tuple Using Tuple Unpacking

def max_min_tuple(tup):
 max_val, min_val = max(tup), min(tup)
 return max_val, min_val

# Test
print(max_min_tuple((5, 2, 9, 1))) # Output: (9, 1)


### 26. Union, Intersection, and Difference of Two Sets

def set_operations(set1, set2):
 union = set1 | set2
 intersection = set1 & set2
 difference = set1 - set2
 return union, intersection, difference

# Test
set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(set_operations(set1, set2)) # Output: ({1, 2, 3, 4, 5}, {3}, {1,2})


### 27. Count Occurrences of an Element in a Tuple

def count_in_tuple(tup, elem):
 return tup.count(elem)

# Test
print(count_in_tuple((1, 2, 2, 3), 2)) # Output: 2


### 28. Symmetric Difference of Two Sets

def symmetric_difference(set1, set2):
 return set1 ^ set2

# Test
set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(symmetric_difference(set1, set2)) # Output: {1, 2, 4, 5}


### 29. Word Frequency in a List of Words

from collections import Counter

def word_frequency(words):
 return dict(Counter(words))

# Test
print(word_frequency(["apple", "banana", "apple", "apple"])) # Output: {’apple’: 3, ’banana’: 1}


### 30. Merge Two Dictionaries

def merge_dicts(dict1, dict2):
 for key in dict2:
  dict1[key] = dict1.get(key, 0) + dict2[key]
 return dict1

# Test
print(merge_dicts({"a": 1, "b": 2}, {"b": 3, "c": 4})) # Output: {’a’: 1, ’b’: 5, ’c’: 4}


### 31. Access Value in a Nested Dictionary

def access_nested_value(d, keys):
 for key in keys:
  d = d.get(key, None)
 if d is None:
  return None
 return d

# Test
nested_dict = {"a": {"b": {"c": 1}}}
print(access_nested_value(nested_dict, ["a", "b", "c"])) # Output: 1


### 32. Sort a Dictionary Based on Values

def sort_dict_by_value(d, reverse=False):
 return dict(sorted(d.items(), key=lambda item: item[1], reverse=reverse))

# Test
print(sort_dict_by_value({"a": 3, "b": 1, "c": 2})) # Output: {’b’: 1, ’c’: 2, ’a’: 3}


### 33. Invert a Dictionary

def invert_dict(d):
 inverted = {}
 for key, value in d.items():
  inverted.setdefault(value, []).append(key)
 return inverted

# Test
print(invert_dict({"a": 1, "b": 2, "c": 1})) # Output: {1: [’a’, ’c’], 2: [’b’]}


### 34. Code to Sort a Dictionary Based on Values (Ascending or Descending)

# This code takes a dictionary as input and sorts it based on the values. You can specify whether the
# sorting should be in ascending or descending order.


def sort_dict_by_value(input_dict, ascending=True):
# Sort the dictionary by value and return the sorted dictionary
 sorted_dict = {k: v for k, v in sorted(input_dict.items(), key=lambda item: item[1], reverse=not ascending)}
 return sorted_dict

# Example usage:
my_dict = {'a': 3, 'b': 1, 'c': 2}

# Sort in ascending order
sorted_dict_asc = sort_dict_by_value(my_dict, ascending=True)
print("Sorted in Ascending Order:", sorted_dict_asc)

# Sort in descending order
sorted_dict_desc = sort_dict_by_value(my_dict, ascending=False)
print("Sorted in Descending Order:", sorted_dict_desc)


### Explanation:
# - The function ‘sort_dict_by_value‘ sorts the dictionary ‘input_dict‘ based on its values.
# - The ‘ascending‘ argument determines whether the sort is in ascending (True) or descending (False)
# order.
# - The ‘sorted‘ function is used with the ‘key‘ argument to sort the dictionary by the value.

### Example Output:

# sorted in Ascending Order: {'b': 1, 'c': 2, 'a': 3}
# sorted in Descending Order: {'a': 3, 'c': 2, 'b': 1}


### 35. Code to Invert a Dictionary (Swap Keys and Values)

# This code inverts a dictionary by swapping its keys and values. If multiple keys have the same value, the
# inverted dictionary stores those keys in a list.


def invert_dict(input_dict):
 inverted_dict = {}

# Iterate through the original dictionary
 for key, value in input_dict.items():
# If the value already exists as a key in the inverted dictionary, append the key to the list
  if value in inverted_dict:
   inverted_dict[value].append(key)
  else:
    inverted_dict[value] = [key]

 return inverted_dict

# Example usage:
my_dict = {'a': 1, 'b': 2, 'c': 1}

# Invert the dictionary
inverted_dict = invert_dict(my_dict)
print("Inverted Dictionary:", inverted_dict)


### Explanation:
# - The function ‘invert_dict‘ swaps the keys and values of ‘input_dict‘.
# - If a value is repeated in the original dictionary (i.e., multiple keys have the same value), the inverted
# dictionary will store all the keys associated with that value as a list.

### Example Output:

#  Inverted Dictionary: {1: ['a', 'c'], 2: ['b']}

```
